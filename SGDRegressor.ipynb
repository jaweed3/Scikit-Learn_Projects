```python
def histogram(df, x, y):
  plt.figure(figsize=(10, 6))
  plt.plot(df[x], df[y], marker='o', linestyle='-', color='b')
  plt.title(f"Histogram of {y}")
  plt.xlabel(x)
  plt.ylabel(y)
  plt.show()

def scatter_plot(df, x, y):
  plt.figure(figsize=(10, 6))
  plt.scatter(df[x], df[y], marker='o', color='b')
  plt.title(f"Scatter Plot of {y} vs {x}")
  plt.xlabel(x)
  plt.ylabel(y)
  plt.show()

def line_plot(df, x, y):
  plt.figure(figsize=(10, 6))
  plt.plot(df[x], df[y], marker='o', linestyle='-', color='b')
  plt.title(f"Line Plot of {y} over {x}")
  plt.xlabel(x)
  plt.ylabel(y)
  plt.show()

def box_plot(df, x, y):
  plt.figure(figsize=(10, 6))
  sns.boxplot(data=df, x=x, y=y)
  plt.title(f"Box Plot of {y} by {x}")
  plt.xlabel(x)
  plt.ylabel(y)
  plt.show()
```

## Importing the libraries for Model building

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score, confusion_matrix
```

### Splitting the dataset

```python
x = selected_data.drop('Calories', axis=1)
y = selected_data['Calories']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

### Standardize the dataset

```python
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

### Use Linear Regression for The Model

```python
clf = LinearRegression()
model = clf.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
y_pred
```

### Result for The Model.

```python
metrics = {
    'MSE': mean_squared_error(y_test, y_pred),
    'R2': r2_score(y_test, y_pred),
}

metrics
```

### Implement K-Fold Validation

```python
scores = cross_val_score(model, x, y, cv=5, scoring='neg_mean_squared_error')
print(f"Cross Validation R2 Scores: {scores}")
print(f"Mean R2 Score: {scores.mean():.4f} (+/- {scores.std() * 2:.4f})")
```

### How to Check Non linear relationship

```python
import numpy as np
import matplotlib.pyplot as plt

# Data
x = np.linspace(-10, 10, 100)
y_quadratic = x**2  # Hubungan kuadratik
y_exponential = np.exp(x)  # Hubungan eksponensial
y_logarithmic = np.log(x[x > 0])  # Hubungan logaritmik (hanya untuk x > 0)

# Plot
plt.figure(figsize=(12, 8))

# Hubungan Kuadratik
plt.subplot(3, 1, 1)
plt.plot(x, y_quadratic, label='y = x^2', color='blue')
plt.title('Hubungan Kuadratik')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

# Hubungan Eksponensial
plt.subplot(3, 1, 2)
plt.plot(x, y_exponential, label='y = e^x', color='green')
plt.title('Hubungan Eksponensial')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

# Hubungan Logaritmik
plt.subplot(3, 1, 3)
plt.plot(x[x > 0], y_logarithmic, label='y = log(x)', color='red')
plt.title('Hubungan Logaritmik')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.legend()

plt.tight_layout()
plt.show()
```

### GridSearchCV Usage

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'alpha' : [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0],
    'learning_rate' : ['constant', 'optimal', 'invscaling', 'adaptive'],
    'eta0' : [0.01, 0.1, 1],
    'max_iter' : [1000, 1500, 2000, 2500, 3000]
}

grid_search = GridSearchCV(regressor, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(x_train, y_train)
best_model = grid_search.best_estimator_
```

### Fitting With The Best Parameter

```python
hist = best_model.fit(x_train, y_train)
```

### Conclusion Matrics

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Prediksi
y_pred = best_model.predict(x_test)

# Hitung metrik dasar
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"MAE: {mae}")
print(f"R2: {r2}")
```

### Prediction VS Actual

### Plotting Learning Curve

```python
from sklearn.model_selection import learning_curve

# Generate learning curve
train_sizes, train_scores, test_scores = learning_curve(
    best_model, x_train, y_train,
    cv=5, n_jobs=-1,
    train_sizes=np.linspace(0.1, 1.0, 10)
)

# Plot learning curve
plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_scores.mean(axis=1), label='Training score')
plt.plot(train_sizes, test_scores.mean(axis=1), label='Cross-validation score')
plt.xlabel('Training Size')
plt.ylabel('Score')
plt.legend(loc='best')
plt.title('Learning Curve')
```

### Last Metrics

```python
def evaluate_model(model, X_train, X_test, y_train, y_test):
    # Prediksi
    y_pred = model.predict(X_test)

    # Metrik evaluasi
    results = {
        'R2 Score': r2_score(y_test, y_pred),
        'MSE': mean_squared_error(y_test, y_pred),
        'RMSE' : np.sqrt(mean_squared_error(y_test, y_pred)),
        'MAE': mean_absolute_error(y_test, y_pred)
    }

    return results

# Gunakan function
evaluation_results = evaluate_model(best_model, x_train, x_test, y_train, y_test)
evaluation_results
```
